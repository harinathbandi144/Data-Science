## 1. Hypotheses Statement
• Null Hypothesis (H₀): The mean weekly operating cost is equal to the theoretical cost.  

H_0: mu = 4000

• Alternative Hypothesis (H₁): The mean weekly operating cost is higher than the theoretical cost.  

H_a: mu > 4000


Formula:

z = x_bar - mu / (sigma/sqrt(n))

• Sample mean (x_bar) = 3050
• Theoretical mean (mu) = 1000 + 5 * 600 = 4000
• Standard deviation of weekly cost (sigma) = 5 * 25 = 125
• Sample size (n) = 25

```python
import numpy as np

# Given data
sigma = 125          # population standard deviation of weekly cost
n = 25               # sample size
x_bar = 3050         # sample mean weekly cost
mu = 4000            # theoretical mean weekly cost

# Calculate Standard Error
SE = sigma / np.sqrt(n)

# Calculate Numerator (difference between sample mean and population mean)
numerator = x_bar - mu

# Calculate Test Statistic (z-score)
z = numerator / SE

# Print results step by step
print("Population Mean (mu):", mu)
print("Sample Mean (x_bar):", x_bar)
print("Population Std Dev (sigma):", sigma)
print("Sample Size (n):", n)
print("Standard Error (SE):", SE)
print("Numerator (x_bar - mu):", numerator)
print("Test Statistic (z):", z)

# Step 4: Compare with critical value at alpha = 0.05 (one-tailed test)
alpha = 0.05
z_critical = 1.645   # critical value for one-tailed test at 5% significance

print("Critical Value (z_critical):", z_critical)


# making a Desicion
if z > z_critical:
    print("Decision: Reject H0 → Evidence costs are higher.")
else:
    print("Decision: Fail to reject H0 → No evidence costs are higher.")
```
