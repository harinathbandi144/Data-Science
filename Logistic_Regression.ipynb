```python
X = df[df.columns[[0,1,2,3,6]]]
Y = df['Clicked on Ad']
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
```

```python
SS_X
```

```python
from sklearn.linear_model import LogisticRegression
LE = LogisticRegression()
```

```python
LE = LE.fit(SS_X, Y)
Y_pred = LE.predict(SS_X)
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

cm = confusion_matrix(Y, Y_pred)
cm
```

```python
accuracy_score(Y,Y_pred)
```

```python
LE.predict_proba(SS_X)
```

```python
df[['Y_prob']] = LE.predict_proba(SS_X)[ : , 1:]
df.head()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
tpr,fpr,dummy = roc_curve(Y,df[['Y_prob']])

rocscore = roc_auc_score(Y,df[['Y_prob']])

print("ROC_Score" , np.round(rocscore,2))
import matplotlib.pyplot as plt

plt.scatter(tpr,fpr)
plt.plot(tpr,fpr,color = 'red')
plt.xlabel('True Positive Rate')
plt.ylabel('False Positive Rate')
```

```python
from sklearn.metrics import log_loss

log_loss = log_loss(Y,df[['Y_prob']])
print(np.round(log_loss,2))
```
