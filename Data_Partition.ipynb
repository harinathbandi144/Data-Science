```python
import pandas as pd

df = pd.read_csv(r'/Users/harinathbandi/Downloads/advertising.csv')
```

```python
df
```

```python
X = df[df.columns[[0,2,3,6]]]
Y = df[['Clicked on Ad']]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler

SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
```

```python
SS_X
```

```python
#Train Test
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(SS_X, Y, test_size= 0.3)

print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
#MODEL Fitting
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train, Y_train)

Y_Pred_train = model.predict(X_train)
Y_Pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
test_acc_score = accuracy_score(Y_test, Y_Pred_test)
train_acc_score = accuracy_score(Y_train, Y_Pred_train)
print("Test",np.round(test_acc_score,2))
print("Train",np.round(train_acc_score,2))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear')
```

```python
# shufflesplit cross validation
from sklearn.model_selection import ShuffleSplit, cross_validate
shuffle_split = ShuffleSplit(n_splits=500, test_size=0.3)

# Perform cross-validation and collect both train and test scores
cv_results = cross_validate(model, SS_X, Y, cv=shuffle_split, scoring='accuracy', return_train_score=True)
```

```python
# Extract train and test scores
train_scores = cv_results['train_score']
test_scores = cv_results['test_score']
```

```python
print("CV-Training Accuracy",np.round(np.mean(train_scores),2))
print("CV-Test Accuracy",np.round(np.mean(test_scores),2))
```

```python
# shufflesplit cross validation
from sklearn.model_selection import KFold, cross_validate
kfold_split = KFold(n_splits=5)

# Perform cross-validation and collect both train and test scores
cv_results = cross_validate(model, SS_X, Y, cv=kfold_split, scoring='accuracy', return_train_score=True)

# Extract train and test scores
train_scores = cv_results['train_score']
test_scores = cv_results['test_score']

print("KFold-Training Accuracy",np.round(np.mean(train_scores),2))
print("KFold-Test Accuracy",np.round(np.mean(test_scores),2))
```
