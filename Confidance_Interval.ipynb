In quality control, sometimes we have to test products in a way that damages or destroys them to check how strong they are. Because each test ruins the item, we can only test a small number of samples instead of the whole batch. To still understand how durable the products are overall, we use something called a confidence interval. This is a statistical method that gives us a range where the true average durability of all products is likely to fall, even though we only tested a few.

#Data
[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
• Sample size: n = 15
• Sample mean: \bar{x} \approx 1.24
• Sample standard deviation: s \approx 0.19

## a. Confidence Interval Using Sample Standard Deviation
• Distribution used: t-distribution (since σ unknown and n small)
• Degrees of freedom: df = n - 1 = 14
• Critical value: t_{0.005,14} \approx 2.977
• Margin of error:  

ME = t \cdot \frac{s}{\sqrt{n}} = 2.977 \cdot \frac{0.19}{\sqrt{15}} \approx 0.15

• Confidence interval:  

(1.24 - 0.15,\; 1.24 + 0.15) = (1.09,\; 1.39)


## b. Confidence Interval Using Known Population Standard Deviation

• Distribution used: z-distribution (since σ known).
• Population standard deviation: \sigma = 0.2.
• Critical value: z_{0.005} \approx 2.576.
• Margin of error:  

ME = z \cdot \frac{\sigma}{\sqrt{n}} = 2.576 \cdot \frac{0.2}{\sqrt{15}} \approx 0.13

• Confidence interval:  

(1.24 - 0.13,\; 1.24 + 0.13) = (1.11,\; 1.37)
