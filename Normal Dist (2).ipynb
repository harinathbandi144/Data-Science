```python
from scipy.stats import norm
nd = norm(167,7.1)  # mean, sd
```

```python
# P(X < 167)  --> -3 to 0
x1 = nd.cdf(167)  # cummulative distributfion function
x1
```

```python
# P(X > 167)  --> 0 to +3
x2 = 1 - x1
x2
```

```python
nd = norm(50000,10000)  # mean, sd
#P(x < 40,000) 
x1 = nd.cdf(40000)  # cummulative distributfion function
x1
```

```python
# What percent of people earning in between 45000 to 60000
# p(45000 < x < 60000)
x3 = nd.cdf(60000) - nd.cdf(45000)
x3
```
