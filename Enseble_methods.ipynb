```python
import pandas as pd
```

```python
df = pd.read_csv('sales.csv')
df
```

```python
df.info()
```

```python

X_cont = df[df.columns[[1,2,3,4,5,7,8]]]
```

```python
#stansderdisation for Integer valiable

from sklearn.preprocessing import StandardScaler

SS = StandardScaler()
SS_X = SS.fit_transform(X_cont)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(df.columns[[1,2,3,4,5,7,8]])
SS_X
```

```python
df.columns
df.info()
```

```python
#Lable encoding for Catogorical variable
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
df['ShelveLoc'] = LE.fit_transform(df['ShelveLoc'])
df['Urban'] = LE.fit_transform(df['Urban'])
df['US'] = LE.fit_transform(df['US'])
df['high'] = LE.fit_transform(df['high'])

df.head()
```

```python
#I need to diff standadised valroable and original variable
df.info()

df_cat = df[df.columns[[6,8,9,10,11]]]

df_cat
```

```python
#Now we will get all continuos variable will come left sife catogorical will come Right side.
#axis = 0 mean row wise, axis = 1 means column wise
df_new = pd.concat([SS_X,df_cat], axis=1)
df_new
```

```python
#we are now catogoring X & Y variable for data partisioning

Y = df_new['high']
X = df_new.drop(['high'],axis=1)

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3)
 
X_train.shape, X_test.shape
```

```python
#above results are changing for every run so, we need to apply cross validation
#apply either shuffle shift and K-FOLD
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='gini')

# Shuffle shift CV
from sklearn.model_selection import ShuffleSplit, cross_validate
shuffle_split = ShuffleSplit(n_splits=200, test_size=0.3, random_state=42)

#performing cross validation to get train test scopres
cv_results = cross_validate(model, X, Y, cv= shuffle_split, scoring='accuracy', return_train_score= True)
import numpy as np

#Extract train test scores
train_score = cv_results['train_score']
test_score = cv_results['test_score']

print(np.round(train_score.mean(),2))
print(np.round(test_score.mean(),2))
```

```python
#we are going for classification bcz we have yes/no target 

# so now I want to remove last 2 nodes

#we are going for classification bcz we have yes/no target 

from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier(criterion='gini',max_depth=7)

DT.fit(X_train,Y_train)

Y_pred_train = DT.predict(X_train)
Y_pres_test = DT.predict(X_test)

from sklearn.metrics import accuracy_score, log_loss

training_accuracy = accuracy_score(Y_train, Y_pred_train)
test_accuracy = accuracy_score(Y_test, Y_pres_test)

training_loss = log_loss(Y_train, Y_pred_train)
test_loss = log_loss(Y_test, Y_pres_test)


import numpy as np
print(np.round(training_accuracy,2))
print(np.round(test_accuracy,2))
print(np.round(training_loss,2))
print(np.round(test_loss,2))
```
